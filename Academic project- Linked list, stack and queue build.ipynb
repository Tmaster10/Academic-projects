```python

class MyLinkedList:

  def __init__(self):
    self.head = None
    self.size = 0
    self.tail = None

  def insertAtBegin(self,new_node):
    new_node.next = self.head
    self.head = new_node
    if  self.tail is None:
      self.tail = new_node
    else:
      self.tail = None
    self.size +=1

  def insertAtEnd(self,new_node):
    if self.head is None:
      self.head = new_node
      self.tail = new_node
      new_node.next = None
      self.size +=1
      return
    else:
      current  = self.head
      while current.next != None:
        current = current.next
      current.next = new_node
      new_node.next = None
      self.size +=1


  def insertAtIndex(self,new_node, index):
    if index < 0 or  index > self.size:
      print("IndexError")
      return
    elif index == 0:
      self.insertAtBegin(new_node)
      return
    elif index == self.size:
      self.insertAtEnd(new_node)
      return
    else:
      current = self.head
      position = 0
      while position < index -1:
        current = current.next
        position +=1
      new_node.next = current.next
      current.next = new_node
      self.size +=1


  def updateNode(self,val,index):
    if index < 0 or index >= self.size:
      print("IndexError")
    else:
      current = self.head
      position = 0
      while current is not None and position < index:
        current = current.next
        position +=1
      if current is not None:
        current.data = val


  def removeAtIndex(self,index):
    if index < 0 or index >= self.size:
       print("IndexError")
    elif index == 0:
      self.head = self.head.next
      self.size -=1
    elif index != 0 and index < self.size:
      current = self.head
      position = 0
      while position < index -1:
        current = current.next
        position +=1
      current.next = current.next.next
      self.size -=1

  def __len__(self):
    return self.size

  def __str__(self):
    current = self.head
    nodes_data = []
    while current is not None:
      nodes_data.append(str(current.data))
      current = current.next
    return("->".join(nodes_data))



```

```python
#task 2

class Stack:

  def  __init__(self):
    self.my_list = MyLinkedList()

  def push(self, value):
    new_node = Node(value)
    self.my_list.insertAtBegin(new_node)

  def pop(self):
    if self.my_list.__len__() == 0:
      raise IndexError("Stack is empty")
    else:
      remove_node = self.my_list.head.data
      self.my_list.removeAtIndex(0)
      return remove_node

  def __len__(self):
    return self.my_list.__len__()

  def __str__(self):
    return str(self.my_list)

  def is_empty(self):
     if self.my_list.__len__() == 0:
      return True
     else:
      return False



class Queue:
  def  __init__(self):
   self.my_list = MyLinkedList()

  def enqueue(self, value):
    new_node = Node(value)
    self.my_list.insertAtEnd(new_node)

  def dequeue(self):
    if self.my_list.__len__() == 0:
      raise IndexError("Queue is empty")
    else:
      remove_node = self.my_list.head.data
      self.my_list.removeAtIndex(0)
      return remove_node

  def __len__(self):
      return self.my_list.__len__()

  def __str__(self):
    return str(self.my_list)

  def is_empty(self):
      if self.my_list.__len__() == 0:
        return True
      else:
        return False



```
