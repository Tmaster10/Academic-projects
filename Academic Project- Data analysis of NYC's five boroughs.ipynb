```python
#Price based on boroughs
boroughs_types = df.groupby('neighbourhood_group')["price"].agg(["mean","max"]).reset_index()

boroughs_types = pd.melt(
    boroughs_types,
    id_vars = ["neighbourhood_group"],
    value_vars = ["mean", "max"],
    var_name = "average_vs_max",
    value_name = "price_value"
)

boroughs_types.head(10)
#most_costly_place_per_boroughs = boroughs_types['price'].max()  # maxmiun price per neighbourhood boruhgs
#average_price_per_boroughs = boroughs_types['price'].mean()

#prices based on neighbours
#neighbours_types = df.groupby('neighbourhood')
#most_costly_place_per_neighbours = neighbours_types['price'].max()  # maxmiun price per neighbourhood boruhgs
#average_price_per_nieghbours = boroughs_types['price'].mean()



```

```python
plt.figure(figsize=(20,10))
AS = sns.barplot(x='neighbourhood_group', y='price_value', hue= 'average_vs_max', data = boroughs_types)
AS.set(xlabel='Neighbourhood borough', ylabel='Prices', title='Average prices and max prices in per New York borough')
for container in AS.containers:
    AS.bar_label(container, fmt = "%.1f", padding =3)
plt.show(AS)
plt.show()

#It seems that when it comes to the highest costing borough group, brookyln and manathann are tied for first for having the highest costly places to rent for an Airbnb 
# While looking into the rental prices, Manathann remain as NO.1 with brooklyn being number 2. 
#Looking at this, you can findind very costly prices no matter the location has even the bronx when the lowest highest pirces are in staten isalnd, if a customer is looking for best prices for an airbnb in BYC
#the best place to find inexpensive housing is the bronx, with staten island coming in second
```

```python
top_5 = df.groupby(['neighbourhood_group', 'neighbourhood'])["price"].agg(["mean","max", 'count']).reset_index()

top_5_by_mean =top_5.sort_values("mean", ascending=False).groupby("neighbourhood_group").head(5)
top_5_by_mean.head(10)

#Sorting by max price trying to get the most neighbours
```

```python
top_5_by_max =top_5.sort_values("max", ascending=False).groupby("neighbourhood_group").head(5)
top_5_by_max.head(10)
#Sorting by mean price (average) trying to get the most neighbours
```

```python
AS = sns.catplot(height = 10, aspect = 1.6, kind = "bar", x='neighbourhood', y='mean', col= 'neighbourhood_group',col_order =['Brooklyn', 'Manhattan', 'Queens', 'Staten Island', 'Bronx'], data = top_5_by_mean,sharex =False, sharey = True, errorbar = None, col_wrap = 1)
AS.set_axis_labels("Neighbourhood", " Mean prices")
AS.set_titles("{col_name}",size=20) 
AS.fig.subplots_adjust(hspace =0.5)
AS.fig.suptitle('Top 5 most costly neighbourhoods per borough based on mean prices',fontsize=20, y=1.03)
AS.fig.tight_layout
for ax in AS.axes.flat:
    ax.xaxis.label.set_size(16)
    ax.yaxis.label.set_size(16)
    ax.tick_params(axis = "x", labelsize = 16, labelrotation = 90)
    ax.tick_params(axis = "y", labelsize = 16)
    for container in ax.containers:
        ax.bar_label(container, fmt = "%.1f", padding =3)

plt.show()

#Graph of the top 5 most costly neighbourhoods per borough based on mean prices
```
