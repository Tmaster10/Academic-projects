```python
class Map:

  def __init__(self):
    self.data = []

  def __setitem__(self,key,value):
    for index, item_dict in enumerate(self.data):
      if item_dict["key"] == key:
        item_dict["value"] = value
        return
    self.data.append({"key": key, "value": value})

  def __getitem__(self, key):
    for index, item_dict in enumerate(self.data):
      if item_dict["key"] == key:
        return item_dict["value"]

    raise KeyError(f'key {key} not found in Map')

  def __delitem__(self, key):
    for index, item_dict in enumerate(self.data):
      if item_dict["key"] == key:
        del self.data[index]
        return
    raise KeyError(f'key {key} not found in Map')

  def __len__(self):
    return len(self.data)

  def __iter__(self):
    for index, item_dict in enumerate(self.data):
      yield item_dict["key"]

  def print_all(self):
    for index, item_dict in enumerate(self.data):
      print(f"Key: {item_dict["key"]}, Value:{item_dict["value"]}")


```

```python
myMap = Map()
```
