```python
graph = {
    "A": ["B", "C"],
    "B":["A", "H"],
    "C":["A", "D", "E","F"],
    "D":["C"],
    "E":["C", "G"],
    "F":["C"],
    "G":["E"],
    "H":["B"],
}
#DFS Implementation
```

```python
visited = { ele: False for ele in graph}
visited
```

```python

def dfs(start):
    if visited[start] == True:
        return
    else:
        print(start)
        visited[start] = True
        neighbours = graph[start]
        for node in neighbours:
            dfs(node)

dfs("A")
```

```python
visited = { ele: False for ele in graph}
def dfs2(start):
    succesors = graph[start]
    if len(succesors) == 0:
        return
    else:
        print(start)
        visited[start] = True
        neighbours = graph[start]
        for node in neighbours:
            dfs2(node)
```

```python
dfs2("C")
```

```python
visited = { ele: False for ele in graph}

def bfs(start):
    queue = []
    cur_position = 0
    queue.append(start)
    #visited[start] = True
    while cur_position < len(queue):
        #print(queue[cur_position])
        if not visited[queue[cur_position]]:
            print(queue[cur_position])
            visited[queue[cur_position]] = True
           
            succesors = graph[queue[cur_position]]
            for node in succesors:
               queue.append(node)

        cur_position += 1
    #print(queue)

bfs("C")
```
