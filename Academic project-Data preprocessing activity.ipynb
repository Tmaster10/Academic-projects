```python
# Convert "Item Purchased" feature (categorical) to a numerical representation. You are not allowed to 
# use OneHotEncoder, LabelEncoder or any similar methods.
df["Item Purchased"] = df["Item Purchased"].astype("category").cat.codes
#df["Item Purchased_codes"] = df["Item Purchased"].cat.codes
```

```python
#Create a heatmap of correlations of all numerical features including 'Item Purchased'


# Select numerical columns for the heatmap
numerical_columns = ['Customer ID', 'Age', 'Item Purchased', 'Purchase Amount (USD)',
       'Review Rating', 'Previous Purchases',]
numerical_df = df[numerical_columns]

# Compute the correlation matrix
corr_matrix = numerical_df.corr().to_numpy()

# Plot the heatmap using Matplotlib
fig, ax = plt.subplots(figsize=(7, 6))  # Adjust figure size
cax = ax.matshow(corr_matrix, cmap='coolwarm')

# Add color bar
plt.colorbar(cax)

# Set axis labels
ax.set_xticks(range(len(numerical_columns)))
ax.set_yticks(range(len(numerical_columns)))
ax.set_xticklabels(numerical_columns, rotation=45, ha="left")
ax.set_yticklabels(numerical_columns)

# Display correlation values inside the heatmap
for i in range(corr_matrix.shape[0]):
    for j in range(corr_matrix.shape[1]):
        ax.text(j, i, f"{corr_matrix[i, j]:.2f}", ha='center', va='center', color='black')

plt.title("Heatmap of Numerical Features")
plt.show()
```
